# Resolución del problema de Monty Hall #

```python
#Importo librerias 
import numpy as np
import random 
import matplotlib.pyplot as plt
```

```python
# Creo una variable y una lista para la opción de quedarse con la puerta que se eligió primero
# Creo otra variable y lista para la opción de cambiar la puerta que se eligió primero.
conviene_quedarse = 0
conviene_cambiar=0
lista_ConvieneQuedarse = []
lista_ConvieneCambiar= []
iteraciones=10000

for repetir in range (iteraciones):
    puertaElegida = random.randint(1,3)
    puertaPremio = random.randint(1,3)
    puertas = [1,2,3]
    puertas.remove(puertaElegida)
    if not(puertaElegida==puertaPremio):
        puertas.remove(puertaPremio)
    puertaRevelada = random.choice(puertas)
    #El conductor revela la puerta en "puertaRevelada"
    if(puertaElegida == puertaPremio):
        conviene_quedarse+=1
        lista_ConvieneQuedarse.append(conviene_quedarse)
    else:
        conviene_cambiar+=1
        lista_ConvieneCambiar.append(conviene_cambiar)
    print("Número de conviene quedarse: ", conviene_quedarse)
    print("Número de conviene cambiar: ", conviene_cambiar)
    
 

        
```

```python
plt.figure(figsize=(16,8))
plt.plot(lista_ConvieneCambiar, label='Conviene cambiar', color='green')
plt.plot(lista_ConvieneQuedarse, label='Conviene quedarse', color='red')
plt.xlabel('Iteración')
plt.ylabel('Aciertos')
plt.legend()
plt.show()
```
